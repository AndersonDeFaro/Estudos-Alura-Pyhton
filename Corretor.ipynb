```python
import nltk
nltk.download('punkt')
```

```python
#Separar palavras de pontuação
palavras_separadas = nltk.tokenize.word_tokenize(texto_exemplo)
print(palavras_separadas)
```

```python
#Função - Separar palavras de pontuação
def separa_palavras(lista_tokens):
  lista_palavras = []
  for token in lista_tokens:
    if token.isalpha():
      lista_palavras.append(token)
       
  return lista_palavras

separa_palavras(palavras_separadas)
```

```python
lista_tokens = nltk.tokenize.word_tokenize(artigos)
lista_palavras = separa_palavras(lista_tokens=lista_tokens)
print('Quantidade de palavras é: ', len(lista_palavras))
```

```python
#frequencia = nltk.FreqDist(lista_normalizada) #Calcula a frequencia das palavras
#frequencia.most_common(10) #Apresenta as 10 palavras mais frequentes

#frequencia['lógica'] #Ocorrência da palavra lógica
#total_palavras = len(lista_normalizada)

#Modelo antigo
#def probabilidade(palavras_geradas):
#  return frequencia[palavras_geradas] / total_palavras

def probabilidade(p_gerada, l_normalizada):
  frequencia = nltk.FreqDist(l_normalizada) # Calcula a frequência das palavras
  return frequencia[p_gerada] / len(l_normalizada)

print('Probabilidade da palavra menionada: ', probabilidade('lógica', lista_normalizada))
```

```python
#probabilidade('logica')
def corretor(palavra, l_normalizada):
  palavras_geradas = gerar_palavras(palavra)
  palavra_correta = max(palavras_geradas, key=probabilidade(p_gerada=p_gerada,l_normalizada=l_normalizada))
  return palavra_correta
```
