```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Lista das criptomoedas
#"BTC-USD", "ETH-USD", "BNB-USD",
tickers = [ "ADA-USD", "SAL-USD", "XRP-USD"]
```

```python
# Data inicial e final
start_date = "2024-01-01"
end_date = "2024-06-30"

# Criar um DataFrame para armazenar os dados
data = pd.DataFrame()
```

```python
# Loop para obter os dados de cada criptomoeda
for ticker in tickers:
    # Obter os dados históricos
    df = yf.download(ticker, start=start_date, end=end_date)
    # Adicionar uma coluna com o nome da criptomoeda
    df['Ticker'] = ticker
    # Concatenar os dados em um único DataFrame
    data = pd.concat([data, df])
```

```python

# Plotar o gráfico
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(data[data['Ticker'] == ticker]['Close'], label=ticker)
plt.legend()
plt.title('Comparação de Preços de Criptomoedas')
plt.xlabel('Data')
plt.ylabel('Preço de Fechamento (USD)')
plt.grid(True)
plt.show()
```
