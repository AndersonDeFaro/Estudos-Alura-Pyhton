## Laços e Loops

```python
def verificar_se_pode_dirigir(idade):
  if idade >= 18:
    print(f'{idade} anos de idade, TEM permissão para dirigir')
  else:
    print(f'{idade} anos de idade, NÂO TEM permissão para dirigir')

for idade in idades:
  verificar_se_pode_dirigir(idade)
```

```python
def loop_verificar_se_pode_dirigir(idades):
  for idade in idades: 
    if idade >= 18:
      print(f'{idade} anos de idade, TEM permissão para dirigir')
    else:
      print(f'{idade} anos de idade, NÂO TEM permissão para dirigir')

loop_verificar_se_pode_dirigir(idades)
```

# Boleano

```python
permissoes = []

def bool_verificar_se_pode_dirigir(idades, permissoes):
  for idade in idades:
    permissoes.append( idade > 18 )

bool_verificar_se_pode_dirigir(idades, permissoes)
permissoes
```

```python
for permissao in permissoes:
  if permissao:
    print('Tem permissao para dirigir')
  else:
    print('NÃO Tem permissao para dirigir')
```

# Imports

```python
from random import randrange, seed

seed(7)
#randrange(0,11)
notas_matematica = []

for nota in range(8):
  notas_matematica.append( randrange(0,11) )

print('Notas de matemarica: ', notas_matematica)
```

# MatPlotLib

```python
import matplotlib.pyplot as plt

x = list(range(1,9))
y = notas_matematica

plt.plot(x, y, marker='o')
plt.title('Notas de Matemática')
plt.xlabel('Provas')
plt.ylabel('Notas')
plt.show()
```
